```python
import pandas as pd
import numpy as np
df=pd.read_excel('E:\Python DEV\Analise_de_Dados\T4.2022.xlsx')
```

```python
def mask(nome_completo):
    """Retorna o primeiro nome e a inicial do sobrenome."""
    partes = nome_completo.split()
    if len(partes) > 1:
        primeiro_nome = partes[0]
        inicial_sobrenome = partes[1][0]  # Primeira letra do sobrenome
        return f"{primeiro_nome} {inicial_sobrenome}."
    return nome_completo  # Caso não haja sobrenome

```

```python
df['Vendedor'] = df['Vendedor'].apply(mask)
df['Cliente']=df['Cliente'].apply(mask)

df
''' Salvar o resultado em um novo arquivo Excel
novo_arquivo_excel = 'nomes_sem_sobrenome.xlsx'
df.to_excel(novo_arquivo_excel, index=False)

print(f"Arquivo salvo como {novo_arquivo_excel}")'''
```
